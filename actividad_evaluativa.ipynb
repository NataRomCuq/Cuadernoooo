```python
import numpy as np  
import matplotlib.pyplot as plt 
from matplotlib.colors import LinearSegmentedColormap

# Definir colores para el degradado en escala de grises
colores = [(0, 'purple'),(0.5, 'cyan'), (1, 'yellow')]  
cmap = LinearSegmentedColormap.from_list('degradado', colores)  

# Crear el degradado lineal en una imagen de 100x100
imagen_degrad = np.linspace(0, 1, 100).reshape(100, 1)  # Gradiente vertical


plt.imshow(imagen_degrad, cmap=cmap, aspect='auto')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

imagen = cv2.imread('Imagen1.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)  
x_inicio, y_inicio, ancho, alto = 660, 600, 175, 235
imagen_recortada = imagen[y_inicio:y_inicio+alto, x_inicio:x_inicio+ancho]
cv2.imwrite("Imagen1_recortada.jpg", cv2.cvtColor(imagen_recortada, cv2.COLOR_RGB2BGR))
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen)

plt.subplot(1, 2, 2)
plt.imshow(imagen_recortada)
plt.show()
```

```python
import numpy as np  
import matplotlib.pyplot as plt 
from matplotlib.colors import LinearSegmentedColormap

# Definir colores para el degradado en escala de grises
colores = [(0, 'black'), (1, 'white')]  
cmap = LinearSegmentedColormap.from_list('degradado', colores)  

# Crear el degradado lineal en una imagen de 100x100
imagen_degrad = np.linspace(0, 1, 100).reshape(100, 1)  # Gradiente vertical


plt.imshow(imagen_degrad, cmap=cmap, aspect='auto')
plt.axis('off')
plt.show()
```
