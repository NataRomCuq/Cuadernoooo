```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

def posterizar(imagen,nivel):
    factor=255/nivel
    image_post=np.round((imagen/factor)*factor)
    return image_post.astype(np.uint8)




bojackbase= cv2.imread('bojackkkk.jpg')
bojackrgb= cv2.cvtColor(bojackbase, cv2.COLOR_BGR2RGB)
bojackgray= cv2.cvtColor(bojackbase, cv2.COLOR_BGR2GRAY)
nivel=3
nueva=posterizar(bojackgray,nivel)
plt.figure(figsize=(20,20))
plt.subplot(2,4,1)
plt.axis('off')
plt.title('bojack normal')
plt.imshow(bojackrgb)


_, umbrellabinario = cv2.threshold(bojackgray, 127, 255, cv2.THRESH_BINARY)
plt.subplot(2,4,2)
plt.axis('off')
plt.title('bojack umbralizado')
plt.imshow(umbrellabinario, cmap='gray')

# Mostrar canal Azul
plt.subplot(2,4,3)
plt.axis('off')
plt.title('bojack en canal azul')
plt.imshow(bojackbase[:,:,0], cmap='Blues')

plt.subplot(2,4,4)
plt.axis('off')
plt.title('bojack en canal verde')
plt.imshow(bojackbase[:,:,1], cmap='Greens')

plt.subplot(2,4,5)
plt.axis('off')
plt.title('bojack en canal rojo')
plt.imshow(bojackbase[:,:,2], cmap='Reds')

plt.figure(figsize=(20,20))
plt.subplot(2,4,6)
plt.axis('off')
plt.title('bojack posterizado')
plt.imshow(nueva, cmap='gray')

```

1 Una imagen digital se puede definir como un conjunto de pixeles que forman una matriz. Estos pixeles pueden tomar distintos valor de acuerdo a color, profundidad, etc.
2 Es un valor de 0 a 255
3 CMYK es un modelo de color basado en cuatro variables: cian, magenta, yellow y uno mas. Se usa principalmente para impresiones. El modelo RGB solo psoee tres variables red, green y blue. Se usa para visualizar las imagenes en pantallas digitales.
