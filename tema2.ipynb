```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer imagen en color
img = cv2.imread('maquiavelo.jpg')  # (alto, ancho, 3 canales)

# Crear kernels
kernel5x5 = np.ones((5,5), np.float32)/25
kernel7x7 = np.ones((7,7), np.float32)/49

# Aplicar convolucin por canal
suavizado = cv2.merge([cv2.filter2D(img[:,:,i], -1, kernel5x5) for i in range(3)])
suavizado2 = cv2.merge([cv2.filter2D(img[:,:,i], -1, kernel7x7) for i in range(3)])

# Mostrar imgenes
plt.figure(figsize=(10,10))

plt.subplot(131)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(132)
plt.imshow(cv2.cvtColor(suavizado, cv2.COLOR_BGR2RGB))
plt.title("Suavizado 5x5")

plt.subplot(133)
plt.imshow(cv2.cvtColor(suavizado2, cv2.COLOR_BGR2RGB))
plt.title("Suavizado 7x7")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('maquiavelo.jpg')

#kernels
kernel_promedio = np.ones((3, 3), np.float32) / 9
kernel_repujado = np.array([[-2, -1, 0],
                          [-1, 1, 1],
                          [0, 1, 2]], np.float32)
kernel_realce = np.array([[0, 0, 0],
                          [-1, 1, 0],
                          [0, 0, 0]], np.float32)
kernel_detbordes = np.array([[0, 1, 0],
                          [1, -4, 1],
                          [0, 1, 0]], np.float32)

kernel_sobel = np.array([[-1, 0, 1],
                          [-2, 0, 2],
                          [-1, 0, 1]], np.float32)


kernel_sharpen = np.array([[1, -2, 1],
                          [-2, 5, -2],
                          [1, -2, 1]], np.float32)

kernel_norte = np.array([[1, 1, 1],
                          [1,-2, 1],
                          [1, -1, -1]], np.float32)

kernel_este = np.array([[-1, 1, 1],
                          [-1,-2, 1],
                          [-1, -1, -1]], np.float32)

def aplicar_kernel_color(img, kernel):
    return cv2.merge([cv2.filter2D(img[:, :, i], -1, kernel) for i in range(3)])

img_promedio = aplicar_kernel_color(img, kernel_promedio)
img_repujado = aplicar_kernel_color(img, kernel_repujado)
img_realce = aplicar_kernel_color(img, kernel_realce)
img_detbordes = aplicar_kernel_color(img, kernel_detbordes)
img_sobel = aplicar_kernel_color(img, kernel_sobel)
img_sharpen = aplicar_kernel_color(img, kernel_sharpen)
img_norte = aplicar_kernel_color(img, kernel_norte)
img_este = aplicar_kernel_color(img, kernel_este)


plt.figure(figsize=(10, 10))

plt.subplot(261)
plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(262)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_promedio, cv2.COLOR_BGR2RGB))
plt.title("Promedio 3x3")

plt.subplot(263)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_bordes, cv2.COLOR_BGR2RGB))
plt.title("Repujado")

plt.subplot(264)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_realce, cv2.COLOR_BGR2RGB))
plt.title("Realce de bordes")

plt.subplot(265)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_detbordes, cv2.COLOR_BGR2RGB))
plt.title("Deteccion de bordes")

plt.subplot(266)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_sobel, cv2.COLOR_BGR2RGB))
plt.title("Sobel")

plt.subplot(267)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_sharpen, cv2.COLOR_BGR2RGB))
plt.title("Sharpen")

plt.subplot(268)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_norte, cv2.COLOR_BGR2RGB))
plt.title("Norte")

plt.subplot(269)
plt.axis('off')
plt.imshow(cv2.cvtColor(img_este, cv2.COLOR_BGR2RGB))
plt.title("Este")


plt.tight_layout()
plt.show()
```
